## Model Training

```python
#Rating
#prepare dataset
prep = apps2[[ 'Rating','Category_new','log_Reviews', 'Size', 'log_Installs','Type', 'Price', 'Content Rating1',  'Recency', 'Android Ver']]
```

```python
prepCategory = pd.get_dummies(prep['Category_new'],columns = 'Category', prefix = 'Category')
prepCategory.head()
```

```python
prepContentrating = pd.get_dummies(prep['Content Rating1'],columns = 'ContentRating', prefix = 'ContentRating')
prepContentrating.head()
```

```python
prepAndroidVer= pd.get_dummies(prep['Android Ver'],columns ='AndroidVer', prefix = 'AndroidVer')
prepAndroidVer.head()
```

```python
prep2 =pd.concat([prep[[ 'Rating','log_Reviews', 'Size', 'log_Installs','Type', 'Price',  'Recency']],prepCategory, prepContentrating,prepAndroidVer], axis =1)
prep2.head()
```

```python
X = prep2.iloc[:,1:-1]
y = prep2['Rating']

X.head()
y.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.33,random_state=42)
```

```python
#using gridsearch to find best parameter for Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

ridge = Ridge()

#define a list of parameters
param_ridge = {'alpha':[0.001, 0.01, 0.1, 1, 10, 100] }

grid_ridge = GridSearchCV(ridge, param_ridge, cv=5, n_jobs=-1, return_train_score = True)
grid_ridge.fit(X_train, y_train)

grid_ridge_train_score = grid_ridge.score(X_train, y_train)
grid_ridge_test_score = grid_ridge.score(X_test, y_test)

print('train score: ', grid_ridge_train_score)
print('test score: ', grid_ridge_test_score)
```

```python
#find best parameters
grid_ridge.best_params_

#scoring results
grid_ridge.cv_results_['mean_test_score']
grid_ridge.cv_results_['mean_train_score']

#predictions using Ridge regression
y_ridge_train_predict = grid_ridge.predict(X_train)
y_ridge_predict = grid_ridge.predict(X_test)
```

```python
from sklearn.metrics.scorer import mean_squared_error as mse
from sklearn.metrics.scorer import mean_absolute_error as mae

mse_ridge_train = mse(y_ridge_train_predict,y_train)
mse_ridge_test = mse(y_ridge_predict,y_test)

mae_ridge_train = mae(y_ridge_train_predict,y_train)
mae_ridge_test = mae(y_ridge_predict,y_test)

print('Train MSE score', mse_ridge_train)
print('Test MSE score',mse_ridge_test )

print('Train MAE score',mae_ridge_train)
print('Test MAE score', mae_ridge_test)
```

```python
#saving results
report_table_regression =[['Ridge', 'alpha=100', grid_ridge_train_score, grid_ridge_test_score,mse_ridge_train, mse_ridge_test, mae_ridge_train, mae_ridge_test]]
report_table_regression
```

```python
#using gridsearch to find best parameter for Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso

lasso = Lasso()

#define a list of parameters
param_lasso = {'alpha':[0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10] }

grid_lasso = GridSearchCV(lasso, param_lasso, cv=5, n_jobs=-1, return_train_score = True)
grid_lasso.fit(X_train, y_train)

grid_lasso_train_score = grid_lasso.score(X_train, y_train)
grid_lasso_test_score = grid_lasso.score(X_test, y_test)

print('train score: ', grid_lasso_train_score)
print('test score: ', grid_lasso_test_score)
```

```python
#predictions using lasso
y_lasso_train_predict = grid_lasso.predict(X_train)
y_lasso_predict = grid_lasso.predict(X_test)

#MSE and MAE
from sklearn.metrics.scorer import mean_squared_error as mse
from sklearn.metrics.scorer import mean_absolute_error as mae

mse_lasso_train = mse(y_lasso_train_predict,y_train)
mse_lasso_test = mse(y_lasso_predict,y_test)

mae_lasso_train = mae(y_lasso_train_predict,y_train)
mae_lasso_test = mae(y_lasso_predict,y_test)

print('Train MSE score', mse_lasso_train)
print('Test MSE score',mse_lasso_test )

print('Train MAE score',mae_lasso_train)
print('Test MAE score', mae_lasso_test)
```
